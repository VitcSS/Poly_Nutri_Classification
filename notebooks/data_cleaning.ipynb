```python
import pandas as pd
```

```python
df_raw =  pd.read_excel("data/raw/survey.xlsx", header = 1)
print(df_raw.columns.to_list())
df_type_1 = df_raw[df_raw.ID_VOLUNT == 1] # Pacientes de Controle
df_type_1 = df_type_1.loc[:, ~(df_type_1.eq(999) | df_type_1.eq(0)).all()].drop(columns=['INICIAIS']).reset_index(drop=True)
df_type_2 = df_raw[df_raw.ID_VOLUNT == 2] # Pacientes com Ovário Policistico
df_type_2 = df_type_2.loc[:, ~(df_type_2.eq(999) | df_type_2.eq(0)).all()].drop(columns=['INICIAIS']).reset_index(drop = True)
```

```python
df_type_1.to_csv("data/staged/Control_P.csv", index = False, sep=";")
df_type_2.to_csv("data/staged/Affected_P.csv", index = False, sep=";")
```
